### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
#adding other dependancies 
import numpy as np
import scipy.stats as sts
import statistics
from matplotlib.ticker import MultipleLocator
import matplotlib.colors as mcolors
```

### Merge the DataFrames

```python
# Combine the data into a single dataset  --- no need to clean sicne we already saw this data was good during the pyber module 
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_type = pyber_data_df.groupby("type").count()["ride_id"]
total_rides_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_citytype = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_citytype 
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = total_fares_citytype / total_rides_type
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fares_citytype / total_drivers_type
avg_fare_per_driver 
```

```python
#  6. Create a PyBer summary DataFrame. 
##first have to flip into DF
pyber_summery_df = pd.DataFrame({"Total Rides": total_rides_type,
                                "Total Drivers": total_drivers_type,
                                "Total Fares": total_fares_citytype,
                                "Average Fare per Ride": avg_fare_per_ride,
                                "Average Fare per Driver": avg_fare_per_driver})
pyber_summery_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summery_df.index.name = None


pyber_summery_df
```

```python
#  8. Format the columns.
pyber_summery_df["Total Fares"] = pyber_summery_df["Total Fares"].map('${:,.2f}'.format)
pyber_summery_df["Average Fare per Ride"] = pyber_summery_df["Average Fare per Ride"].map('${:,.2f}'.format)
pyber_summery_df["Average Fare per Driver"] = pyber_summery_df["Average Fare per Driver"].map('${:,.2f}'.format)

pyber_summery_df.head()

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
date_sum_df = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
date_sum_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
date_sum_df = date_sum_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
date_sum_df  = date_sum_df .pivot(index="date", columns="type", values="fare")

date_sum_df .head()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
jan_to_may_fares = date_sum_df.loc['2019-01-01':'2019-04-29']
jan_to_may_fares
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
jan_to_may_fares.index = pd.to_datetime(jan_to_may_fares.index)
jan_to_may_fares.head()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
jan_to_may_fares.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
total_fare_per_week = jan_to_may_fares.resample("W").sum()
total_fare_per_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style

#use object oriented interface method and df.plot() method
ax = total_fare_per_week.plot()

ax.plot()
ax.set_title("Total Fare by City Type")
ax.set_ylabel("FAre ($USD)")
ax.set_xlabel("2019 by months (Jan1-May")


# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()

```
